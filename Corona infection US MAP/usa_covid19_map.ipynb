```python
import pandas as pd
import folium
from folium import plugins
import numpy as np
from folium.plugins import HeatMap
```

```python
dataset=pd.read_excel('usa_covid19.xlsx')
dataset.head()
```

```python
place =dataset[["lat","long"]] # to import the latitude and longitude
place.head()
```

```python
place=place.values.tolist()
```

```python
usa_map = [ 38.500000,-98.000000] 
covid_map = folium.Map(location = usa_map, 
                            tiles = 'Stamen Toner', 
                            zoom_start = 3.8)
```

```python
def mark(i,color):
    # Function to create marker
    folium.Marker(
        location=point, # To give Latitude and Longitude stored in point
        popup=dataset["state"][i] + " " + str(dataset["cases"][i]),  # Assigning label to the marker as State_name Confirmed covid cases
        icon=folium.Icon(
            color=color,  # Assigning color
            icon='tint',
            icon_color='white'
        )
    ).add_to(covid_map)
    
```

```python
i=0
for point in place:
    if dataset['cases'][i]>50000:
        # Alert
        mark(i,'darkred')
    elif dataset['cases'][i]>40000 and dataset['cases'][i]<=50000:
        # High Risk
        mark(i,'red')
    elif dataset['cases'][i]>30000 and dataset['cases'][i]<=40000:
        # Risk
        mark(i,'lightred')
    elif dataset['cases'][i]>20000 and dataset['cases'][i]<=30000:
        # Average
        mark(i,'orange')
    elif dataset['cases'][i]>10000 and dataset['cases'][i]<=20000:
        # Below average
        mark(i,'lightgreen')
    elif dataset['cases'][i]>0 and dataset['cases'][i]<=10000:
        # Below average
        mark(i,'green')
    i+=1
```
