```python
import requests
from bs4 import BeautifulSoup
```

```python
page_urls = ['https://www.indeed.com/jobs?q=data+scientist',
             'https://www.indeed.com/jobs?q=data+scientist&start=10',
             'https://www.indeed.com/jobs?q=data+scientist&start=20',
             'https://www.indeed.com/jobs?q=data+scientist&start=30',
             'https://www.indeed.com/jobs?q=data+scientist&start=40'
            ]

```

```python
title = []
company = []
desc = []
location = []
rating=[]
salary= []
for url in page_urls:
    response = requests.get(url)
    if response.status_code !=200:
        continue
    else:
        html_code =response.content
        soup = BeautifulSoup(html_code, 'lxml')
        result_col = soup.find('td', {'id': 'resultsCol'} )
        for i in result_col.find_all('h2',{'class': 'title'}):
            title.append(i.get_text(strip=True))
        for j in result_col.find_all('span',{'class': 'company'}):
            company.append(j.get_text(strip=True))
        for a in result_col.find_all('div',{'class': 'summary'}):
            desc.append(a.get_text(strip=True))
        for b in result_col.find_all('div',{'class': 'location accessible-contrast-color-location'}):
            location.append(b.get_text(strip=True))
        for c in result_col.find_all('span',{'class': 'ratingsContent'}):
            rating.append(c.get_text(strip=True))
        for d in result_col.find_all('div',{'class': 'salarySnippet'}):
            salary.append(d.get_text(strip=True))
```
